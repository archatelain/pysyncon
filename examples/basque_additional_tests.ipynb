```python
import pandas as pd
import numpy as np
from pysyncon import Dataprep, Synth
from pysyncon.utils import PlaceboTest
```

```python
df = pd.read_csv("../data/basque.csv")
```

```python
first_predictor = df.query('year==1955')[['regionname', 'gdpcap']].rename(columns={'gdpcap': 'first_gdpcap'})
df = df.merge(first_predictor, on='regionname', how='left')
```

```python
special_list = []
for year in range(1956, 1971):
    special_list.append(
        ('gdpcap', range(year, year+1), 'mean')
    )
```

```python
dataprep = Dataprep(
    df, 
    predictors=['first_gdpcap'],
    predictors_op='mean',
    dependent="gdpcap",
    unit_variable="regionname",
    time_variable="year",
    treatment_identifier="Basque Country (Pais Vasco)",
    controls_identifier=[
        "Spain (Espana)",
        "Andalucia",
        "Aragon",
        "Principado De Asturias",
        "Baleares (Islas)",
        "Canarias",
        "Cantabria",
        "Castilla Y Leon",
        "Castilla-La Mancha",
        "Cataluna",
        "Comunidad Valenciana",
        "Extremadura",
        "Galicia",
        "Madrid (Comunidad De)",
        "Murcia (Region de)",
        "Navarra (Comunidad Foral De)",
        "Rioja (La)",
    ],
    time_predictors_prior=range(1955, 1970),
    time_optimize_ssr=range(1955, 1970),
    special_predictors=special_list
)
```

```python
custom_V = np.ones(len(range(1955, 1971)))
```

```python
synth = Synth()
synth.fit(dataprep=dataprep, custom_V=custom_V)
```
